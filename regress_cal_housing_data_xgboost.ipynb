```python
## import necessary libraries
import os
from sklearn.datasets import fetch_california_housing
from xgboost_prediction.build_predict_model import predict_regression_model
```

```python
# Load the Iris dataset
X, y = fetch_california_housing(return_X_y=True)
print(X.shape)
```

```python
params = {
    'objective': 'reg:squarederror',  # Objective for regression problems
    'learning_rate': 0.1,             # Controls the step size for updates
    'max_depth': 5,                   # Limits the tree depth to prevent overfitting
    'eval_metric': 'rmse',            # Root Mean Square Error (or use 'mae' for Mean Absolute Error)
    'subsample': 0.8,                 # Use 80% of data for training each tree (controls overfitting)
    'colsample_bytree': 0.8           # Use 80% of features for training each tree (also controls overfitting)
}
```

```python
# Perform predictions
model_path = 'iris_models/cal_housing_model_features.json'
fig = predict_regression_model(X, y, model_path, feature_count=X.shape[-1], params=params, output_shap_plot=True)
```
